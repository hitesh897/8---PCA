```python
# Importing Libraries
import pandas as pd 
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale 
```

```python
# Import Dataset
wine = pd.read_csv("wine.csv")
wine.describe()
wine.head()
```

```python
wine['Type'].value_counts()
```

```python
wine1 = wine.iloc[:,1:]
wine1
```

```python
wine1.shape
```

```python
wine1.info()
```

```python
# Converting data to numpy array
wine_ary = wine1.values
wine_ary
```

```python
# Normalizing the Numerical data
wine_norm = scale(wine_ary)
wine_norm
```

###  PCA Implementation

```python
# Applying PCA Fit Transform to dataset
pca = PCA(n_components = 13)

wine_pca = pca.fit_transform(wine_norm)
wine_pca
```

```python
# PCA Components matrix or covariance matrix

pca.components_
```

```python
# The amount of variance each PCA has

var = pca.explained_variance_ratio_
var
```

```python
# Cummulative variance of each PCA
var1=np.cumsum(np.round(var,4)*100)
var1
```

```python
# Variance plot for PCA components obtained 
plt.plot(var1,color="red")
```

```python
#Final Dataframe
finalDf = pd.concat([wine['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis = 1)
finalDf
```

```python
# Visualization of PCA's
fig = plt.figure(figsize=(16,12))
sns.scatterplot(data=finalDf)
```

#### Performing Hierarchical clustering using first 3 principal component scores 

```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
# Creating Dendrograms using our Normalized data
plt.figure(figsize=(10,8))
dendrograms = sch.dendrogram(sch.linkage(wine_norm,'complete'))
```

```python
# create clusters
hclusters = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'ward')
hclusters
```

```python
y = pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y
```

```python
y['clustersid'].value_counts()
```

```python
# Adding Clusters column in Dataset
wine2 = wine.copy()
wine2['clustersid']=hclusters.labels_
wine2
```

###  K-means clustering

```python
from sklearn.cluster import KMeans
```

```python
# We already have Normalized data from PCA
# We will use Elbow method to find number of clusters to be created
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

```python
# Elbow curve
wcss = []
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i,random_state=0)
    kmeans.fit(wine_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 6), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

####  We will build cluster using k=3

```python
#Build Cluster algorithm
clusters_new = KMeans(3, random_state=30)
clusters_new.fit(wine_norm)
```

```python
clusters_new.labels_
```

```python
# Adding Clusters column in the Dataset
wine3 = wine.copy()
wine3['clusterids']=clusters_new.labels_
wine3
```
